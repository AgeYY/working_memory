# Introduction

This project explores a unified model for short-term memory. The to-be-memorized variable can draw values from any type of distributions. We also study how the prior distribution of to-be-memorized variable effect the representation and computation of short-term memory.

To achieve our goal, we trained RNN to perform short-term memory task. Next we compare the behavior and neural activity of the RNN to the brain. If they are similar, then the model the RNN used is also likely to be used by the brain. We analyzed the model of RNN mainly by principal component analysis (PCA). Finally, we offer top-down explanation of why the RNN would use such model.

# Model Training

We trained RNN to do delayed-response task. The task has 5 epochs: fixation, perception, delay, go, response. In the perception epoch, the value of the to-be-memorized variable would be drawn from certain biased prior distribution. The RNN need to remember this value, and then reproduce this value after short period of delay.

We provide two ways for training the RNN. _bin/train\_local_.py can train the model locally. Another way is _bin/train_cluster.py_ and _bin/train\_cluster.sh_, which includes MPI4Py for parallel training. Training one RNN locally takes 10 mins.

The trained RNNs are provided by link here.

```python
# training serveral configuration on cluster by mpi4py
# the default trtaning task are 32 and cones, but you can identify it by passing the argument
import context
from core.train_batch import train_model
import sys
import os

sys.path.append(os.getcwd())

#################### Parameters
is_cuda = True
rule_name = "color_reproduction_delay_unit" # RNN types: can be "color_reproduction_delay_unit", or 'color_reproduction_delay_tri'. See explanation in default.py
n_models = 1

model_base = '../core/jupyter_model/'

hp_replace = {'prod_interval': [0, 1000], 'l2_jac': -1, 'sigma_rec': 0.2, 'sigma_x': 0.2, 'num_unit': 12, \
              'n_input': 13, 'n_output': 12, 'n_rnn': 256, 'bias_method': 'vonmises', 'sig': 25, 'stop_color_error': 1, \
              'stop_noise_color_error': 30, 'stop_delta_color_error': 30, 'min_trials': 2e5, \
              'bias_centers': [40., 130., 220., 310.], 'center_prob': 0.5,
             } # We do not set threshold for stopping training.
```

Show the prior distribution of the to-be-memorized variable

```python
from core.color_input import input_color_dist
import matplotlib.pyplot as plt
import numpy as np

color_samples = np.linspace(0, 360, 360)
center_p_input = input_color_dist(color_samples, hp_replace['bias_centers'], hp_replace['bias_method'], sig=hp_replace['sig'], center_prob=hp_replace['center_prob'])

fig = plt.figure(0, figsize=(5, 5))
ax = fig.add_subplot(111)

ax.plot(color_samples, center_p_input)
ax.set_ylim([0, 0.01])
plt.title('prior distribution of the to-be-memorized variable')
plt.show()
```

Start training. We employed progressive training here. In the first training stage, we turned off the noise. The delay period length is 0. The prior distribution of to-be-memorized variable is uniform. In the second stage, the delay period is randomly chosen from interval \[0, 1000\]. Third, delay period is \[0, 2000\]. Next, we add noise and regularization. Finally, we use biased prior distribution which is controled by parameters: sig, bias_method, center_prob (if bias_method is delta). This parameters are defined in hp\_replace

```python
#################### Training
for i in range(n_models):
    ## We use progressive training
    # step 1: prod_interval = [0, 0], no noise, no penalty, uniform prior distribution
    hp_replace_temp = hp_replace.copy()

    hp_replace_temp['prod_interval'] = [0, 0]
    hp_replace_temp['sigma_rec'] = 0.
    hp_replace_temp['sigma_x'] = 0
    hp_replace_temp['l2_jac'] = -0.1 # negative means no penalty
    hp_replace_temp['l2_weight'] = -0.1
    hp_replace_temp['l2_firing_rate'] = -0.1
    hp_replace_temp['bias_method'] = 'uniform'
    hp_replace_temp['learning_rate'] = 0.0005 # In the first step we need larger learning rate, to find the solution.

    cost = train_model(rule_name, i, hp_replace=hp_replace_temp, is_cuda=is_cuda, final_name='progressive_' + str(0), model_base=model_base)

    # step 2: prod_interval = [0, 1000], adjust the learning rate
    hp_replace_temp['prod_interval'][1] = hp_replace_temp['prod_interval'][1] + 1000
    hp_replace_temp['learning_rate'] = 0.00005
    cost = train_model(rule_name, i, hp_replace=hp_replace_temp, is_cuda=is_cuda, final_name='progressive_' + str(1), model_base=model_base)
    
    # step 3: with noise
    hp_replace_temp = hp_replace.copy()
    hp_replace_temp['bias_method'] = 'uniform'

    cost = train_model(rule_name, i, hp_replace=hp_replace_temp, is_cuda=is_cuda, final_name='noise', model_base=model_base)

    # step 4: original hp_replace
    cost = train_model(rule_name, i, hp_replace=hp_replace, is_cuda=is_cuda, final_name='noise_delta', model_base=model_base)


#################### clear failed models, i.e. trials exceed max_trials
from core.model_seletor import Model_seletor

try:
    max_trials = hp_replace['max_trials']
except:
    max_trials = 1e6

ms = Model_seletor()
ms.delete(model_base + rule_name)
```

# Neural Encoding

From the manifold, we can see that different input colors during the perception epoch drive the neural state into different angular positions. Therefore the information of color may be encoded by different angular positions on the PCA plane. To verify this hyperthesis, we built and RNN Decoder which decodes a neural state to its encoded color. Specificly, we have a RNN trained to perform the delayed-response task. If we want to know what is the color encoded by neural state A, we only need to preset the RNN's state to the neural state A, then immediately run the RNN through the go and response epoch. The report color during the response epoch would be the encoded color of neural state A. Hence this process maps a neural state to a encoded color. We define this process as RNN Decoder. We use RNN Decoder to decode neural states on the first-second priciple component plane.

```python
from core.rnn_decoder import RNN_decoder
from core.manifold.fix_point import Hidden0_helper # help meshing points in the neural state space
from core.color_manager import Degree_color # for visualization
from sklearn.decomposition import PCA
```

Firstly, we find out the first-second principle components plane by fitting neural activities during the delay

```python
##### fit data to find the pca plane
n_components = 2
pca = PCA(n_components=n_components)
pca.fit(sub.fir_rate[sub.epochs['interval'][1]]);
```

Knowing the principle components planes, next we decode every points on the plane by RNN Decoder. Specificly, we firstly sample mesh points on the principle components plane, then use inverse pca to map them back to the original, high dimensional neural states. Than decode these neural states by RNN Decoder.

```python
##### Mesh points on the pca plane
xlim=[-50, 50]; ylim=[-50, 50]; edge_batch_size=200;

hidden_size = sub.state.shape[-1] # dimensionality of the original neural states, which is also the number of neurons
hhelper = Hidden0_helper(hidden_size=hidden_size)
hidden0_grid_pca, hidden0_grid = hhelper.mesh_pca_plane(sub, xlim, ylim, edge_batch_size) # mesh points on the pca plane.

hidden0_grid_rate = np.tanh(hidden0_grid) + 1 # convert to firing rate
hidden0_grid_pca = pca.transform(hidden0_grid_rate)

##### Decode mesh points
rnn_de = RNN_decoder() # RNN decoder
rnn_de.read_rnn_agent(sub)

report_color_grid = rnn_de.decode(hidden0_grid)
```

hidden0_grid_pca is the neural states' on the pca plane. report_color_grid is the corresponding decoded color. Now visualize them

```python
deg_color = Degree_color()
colors_grid = deg_color.out_color(report_color_grid, fmat='RGBA') # use RGBA color for every reported color

fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

ax.scatter(hidden0_grid_pca[:, 0], hidden0_grid_pca[:, 1], c=colors_grid, alpha=1, s=30)
plt.title('first-second principle components plane')
plt.show()
```

Color in this figure represent different variable values. Indeed, it looks that different angles on the pca plane encodes different variable values. We quantify the relationship between the angle and the variable value. To do this, we sample points with angles phi in the pca plane. Next we use inverse pca to map them back to the original neural states. Finally we decode the neural states with RNN Decoder

```python

##### state in the hidimensional space and pca plane
hidden_size = sub.state.shape[-1]
hhelper = Hidden0_helper(hidden_size=hidden_size)
# We save the neural states which at the end of delay epoch, i.e. circle points in the above interval figure. So hidden0_ring 
# is a matrix of [n_batch, n_neuron], where n_batch is the number of input colors in one batch, i.e. the length of input_color,
# n_neuron is the dimension of a neural state.
hidden0_ring = hhelper.noisy_ring(sub, sigma_init=0)
hidden0_ring_pca = pca.transform(hidden0_ring) # then we map these end neural states into principle components plane. The
# shape should be a ring.
    
##### decode states from high dimesional space
rnn_de = RNN_decoder()
rnn_de.read_rnn_agent(sub)
    
report_color_ring = rnn_de.decode(hidden0_ring)
deg_pca = np.arctan2(hidden0_ring_pca[:, 1], hidden0_ring_pca[:, 0]) # calculate the angle of hidden0_ring_pca
deg_pca = np.mod(deg_pca, 2*np.pi) / 2.0 / np.pi * 360.0
```
